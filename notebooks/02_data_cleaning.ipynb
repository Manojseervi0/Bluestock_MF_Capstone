```python
import pandas as pd

nav_df = pd.read_csv("../data/raw/02_nav_history.csv")

print("Shape:", nav_df.shape)
print("\nData Types:")
print(nav_df.dtypes)

print("\nMissing Values:")
print(nav_df.isnull().sum())

print("\nDuplicate Rows:")
print(nav_df.duplicated().sum())

print("\nNAV <= 0:")
print((nav_df["nav"] <= 0).sum())
```

```python
nav_df["date"] = pd.to_datetime(nav_df["date"])

nav_df = nav_df.sort_values(
    ["amfi_code", "date"]
)

print(nav_df.dtypes)

print("\nFirst 5 rows:")
print(nav_df.head())
```

```python
print("Unique Funds:")
print(nav_df["amfi_code"].nunique())

print("\nDate Range:")
print(nav_df["date"].min())
print(nav_df["date"].max())

print("\nRecords Per Fund:")
print(nav_df.groupby("amfi_code").size().head())
```

```python
nav_df.to_csv(
    "../data/processed/02_nav_history_clean.csv",
    index=False
)

print("Clean NAV file saved successfully")
```

```python
txn_df = pd.read_csv("../data/raw/08_investor_transactions.csv")

print("Shape:", txn_df.shape)

print("\nMissing Values:")
print(txn_df.isnull().sum())

print("\nTransaction Types:")
print(txn_df["transaction_type"].value_counts())

print("\nKYC Status:")
print(txn_df["kyc_status"].value_counts())

print("\nAmount <= 0:")
print((txn_df["amount_inr"] <= 0).sum())
```

```python
perf_df = pd.read_csv("../data/raw/07_scheme_performance.csv")

print("Shape:", perf_df.shape)

print("\nMissing Values:")
print(perf_df.isnull().sum())

print("\nData Types:")
print(perf_df.dtypes)

print("\nExpense Ratio Range:")
print(
    perf_df["expense_ratio_pct"].min(),
    "to",
    perf_df["expense_ratio_pct"].max()
)
```

```python
return_cols = [
    "return_1yr_pct",
    "return_3yr_pct",
    "return_5yr_pct"
]

for col in return_cols:
    print(f"\n{col}")
    print(pd.to_numeric(perf_df[col], errors="coerce").isnull().sum())
```

```python
perf_df.to_csv(
    "../data/processed/07_scheme_performance_clean.csv",
    index=False
)

print("Clean performance file saved successfully")
```
